```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
marks_john = [265.3, 227.5]
marks_sam = [339.1, 308]

# create plot
fig, ax = plt.subplots()
bar_width = 0.35
X = np.arange(2)

p1 = plt.bar(X, marks_john, bar_width, color='b',
label='With Smart Units Extension')

# The bar of second plot starts where the first bar ends
p2 = plt.bar(X + bar_width, marks_sam, bar_width,
color='g',
label='Without Smart Units Extension')

plt.xlabel('')
plt.ylabel('Sec')
plt.title('Comparision of Performance with and without Smart Units Extension')
plt.xticks(X + (bar_width/2) , ("Mean", "Median"))
plt.legend()
plt.rcParams['figure.figsize'] = (10, 6)
plt.show()
```
